#### Create overlapped representation

```python
from copy import deepcopy

def create_overalaped_representation2(manager, representation_overlap, sequence_overlap):
    x = deepcopy(manager.canonical_activity_representation)

    to_modify = int(representation_overlap * len(x[0]))
    sequence_size = int(0.5 * len(x))
    sequence_overlap_size = int(sequence_overlap * sequence_size)
    start_point = int(0.5 * sequence_size + sequence_size - np.floor(sequence_overlap_size/ 2.0))
    end_point = start_point + sequence_overlap_size

    for sequence_index in range(start_point, end_point):
        pattern = x[sequence_index]
        pattern[:to_modify] = manager.canonical_activity_representation[sequence_index - start_point][:to_modify]

    return x
```

#### Calculate total recall

```python
def calculate_total_recall(manager, representation, patterns_per_sequence, T_persistence=0.050, remove=remove):
    manager.set_persistent_time_with_adaptation_gain(T_persistence=T_persistence)

    T_cue = manager.nn.tau_s
    T_recall = T_persistence * patterns_per_sequence + T_cue
    end1 = patterns_per_sequence - 1
    end2 = 2 * patterns_per_sequence - 1

    nr1 = representation.network_representation[:patterns_per_sequence]
    nr2 = representation.network_representation[patterns_per_sequence:]


    I_cue = nr1[0]
    nr = nr1
    aux1 = calculate_recall_quantities(manager, nr, T_recall, T_cue, remove=remove, reset=True, empty_history=True)
    success1, pattern_sequence1, persistent_times1, timings1 = aux1
    sequence_recall1 = evaluate_disambiguation_success(end=end1, wrong_end=end2, pattern_sequence=pattern_sequence1)

    I_cue = nr2[0]
    nr = nr2
    aux2 = calculate_recall_quantities(manager, nr, T_recall, T_cue, remove=remove, reset=True, empty_history=True)
    success2, pattern_sequence2, persistent_times2, timings2 = aux2
    sequence_recall2 = evaluate_disambiguation_success(end=end2, wrong_end=end1, pattern_sequence=pattern_sequence2)
    
    total_success = 0.5 * success1 + 0.5 * success2
    
    results_dic = {}
    results_dic['success'] = total_success
    results_dic['success_ending'] = 0.5 * (sequence_recall1 + sequence_recall2)
    results_dic['ending_success1'] = sequence_recall1
    results_dic['ending_success2'] = sequence_recall2
    results_dic['recall_quantities1'] = aux1
    results_dic['recall_quantities2'] = aux2
    
    return results_dic
```

#### Calculate total recall trials

```python
def calculate_total_recall_trials(trials, manager, representation, patterns_per_sequence, 
                                   T_persistence=0.100, remove=remove):

    ending_success_vector1 = []
    ending_success_vector2 = []
    success1_vector = []
    success2_vector = []
    sequences1 = []
    sequences2 = []

    for i in range(trials):
        recall_dic = calculate_total_recall(manager, representation, patterns_per_sequence, 
                                            T_persistence=T_persistence, remove=remove)
        ending_success1 = recall_dic['ending_success1']
        ending_success2 = recall_dic['ending_success2']
        aux1 = recall_dic['recall_quantities1']
        success1, pattern_sequence1, persistent_times1, timings1 = aux1
        aux2 = recall_dic['recall_quantities2']
        success2, pattern_sequence2, persistent_times2, timings2 = aux2

        ending_success_vector1.append(ending_success1)
        ending_success_vector2.append(ending_success2)
        success1_vector.append(success1)
        success2_vector.append(success2)
        sequences1.append(pattern_sequence1)
        sequences2.append(pattern_sequence2)

    total_success = [0.5 * (x + y) for (x, y) in zip(success1_vector, success2_vector)]
    total_success_ending = [0.5 * (x + y) for (x, y) in zip(ending_success_vector1, ending_success_vector2)]

    results_dic = {'ending_success1':np.array(ending_success_vector1), 
                   'ending_success2':np.array(ending_success_vector2),
                   'success1':np.array(success1_vector), 'success2':np.array(success2_vector),
                   'sequences1':sequences1, 'sequences2':sequences2, 
                   'success':np.array(total_success), 'success_ending':np.array(total_success_ending)}
    
    return results_dic
```

#### Calculate overlap success

```python
def calculate_overlap_success(representation_overlap=0.5, sequence_overlap=0.1, minicolumns=20, hypercolumns=10, 
                              tt=0.100, ipi=0, tau_z_pre=0.025, tau_z_post=0.005, sigma_out=0.0, T_persistence=0.100,
                             remove=remove):
    
    strict_maximum = True

    g_a = 1.0
    g_I = 2.0
    tau_a = 0.250
    G = 1.0
    sigma_out = sigma_out
    tau_s = 0.010
    tau_z_pre = tau_z_pre
    tau_z_post = tau_z_post

    hypercolumns = hypercolumns
    minicolumns = minicolumns
    n_patterns = minicolumns
    patterns_per_sequence = int(n_patterns / 2)
    representation_overlap = representation_overlap
    sequence_overlap = sequence_overlap

    # Training protocol
    training_times_base = tt
    training_times = [training_times_base for i in range(n_patterns)]
    ipi_base = ipi
    inter_pulse_intervals = [ipi_base for i in range(n_patterns)]
    inter_sequence_interval = 1.0
    resting_time = 1.0
    epochs = 1

    # Manager properties
    values_to_save = ['o']


    # Neural Network
    nn = Network(hypercolumns, minicolumns, G=G, tau_s=tau_s, tau_z_pre=tau_z_pre, tau_z_post=tau_z_post,
                     tau_a=tau_a, g_a=g_a, g_I=g_I, sigma_out=sigma_out, epsilon=epsilon, prng=np.random,
                     strict_maximum=strict_maximum, perfect=False, normalized_currents=True)


    # Build the manager
    manager = NetworkManager(nn=nn, dt=dt, values_to_save=values_to_save)
    # Build the representation and put a time between them
    activity_representation = create_overalaped_representation(manager, representation_overlap, sequence_overlap)
    representation = PatternsRepresentation(activity_representation, minicolumns=minicolumns)
    inter_pulse_intervals[patterns_per_sequence - 1] = inter_sequence_interval
    # Build the protocol
    protocol = Protocol()
    protocol.simple_protocol(representation, training_times=training_times, inter_pulse_intervals=inter_pulse_intervals,
                        inter_sequence_interval=inter_sequence_interval, epochs=epochs, resting_time=resting_time)

    # Run the protocol
    timed_input = manager.run_network_protocol_offline(protocol=protocol)

    recall_dic = calculate_total_recall(manager, representation, patterns_per_sequence, 
                                        T_persistence=T_persistence, remove=remove)
    
    return recall_dic, manager
```

#### Overlap success trials

```python
def calculate_overlap_success_trials(representation_overlap=0.5, sequence_overlap=0.1, minicolumns=20, 
                                     hypercolumns=10, tt=0.100, ipi=0, tau_z_pre=0.025, tau_z_post=0.005, 
                                     sigma_out=0.0, T_persistence=0.100, trials=10, remove=remove):
    
    strict_maximum = True

    g_a = 1.0
    g_I = 2.0
    tau_a = 0.250
    G = 1.0
    sigma_out = sigma_out
    tau_s = 0.010
    tau_z_pre = tau_z_pre
    tau_z_post = tau_z_post

    hypercolumns = hypercolumns
    minicolumns = minicolumns
    n_patterns = minicolumns
    patterns_per_sequence = int(n_patterns / 2)
    representation_overlap = representation_overlap
    sequence_overlap = sequence_overlap

    # Training protocol
    training_times_base = tt
    training_times = [training_times_base for i in range(n_patterns)]
    ipi_base = ipi
    inter_pulse_intervals = [ipi_base for i in range(n_patterns)]
    inter_sequence_interval = 1.0
    resting_time = 1.0
    epochs = 1

    # Manager properties
    values_to_save = ['o']


    # Neural Network
    nn = Network(hypercolumns, minicolumns, G=G, tau_s=tau_s, tau_z_pre=tau_z_pre, tau_z_post=tau_z_post,
                     tau_a=tau_a, g_a=g_a, g_I=g_I, sigma_out=sigma_out, epsilon=epsilon, prng=np.random,
                     strict_maximum=strict_maximum, perfect=False, normalized_currents=True)


    # Build the manager
    manager = NetworkManager(nn=nn, dt=dt, values_to_save=values_to_save)
    # Build the representation

    activity_representation = create_overalaped_representation(manager, representation_overlap, sequence_overlap)

    representation = PatternsRepresentation(activity_representation, minicolumns=minicolumns)
    inter_pulse_intervals[patterns_per_sequence - 1] = inter_sequence_interval
    # Build the protocol
    protocol = Protocol()
    protocol.simple_protocol(representation, training_times=training_times, inter_pulse_intervals=inter_pulse_intervals,
                        inter_sequence_interval=inter_sequence_interval, epochs=epochs, resting_time=resting_time)

    # Run the protocol
    timed_input = manager.run_network_protocol_offline(protocol=protocol)
    
    results_dic = calculate_total_recall_trials(trials, manager, representation, patterns_per_sequence, 
                                                 T_persistence=T_persistence, remove=remove)
    
    return results_dic
```

#### Sigma sweep

```python
def run_sigma_sweep(sigma_number, trials, sigma_max, manager, representation, patterns_per_sequence,
                    T_persistence=0.100, remove=remove, cue_times=2):
    
    sigma_vector = np.linspace(0, sigma_max, num=sigma_number)
    successes_collection = np.zeros((sigma_number, trials))

    for index, sigma in enumerate(sigma_vector):
        manager.nn.sigma_out = sigma
        manager.nn.sigma_in = sigma * np.sqrt(2 / manager.nn.tau_s)
        aux = calculate_total_recall_trials(trials, manager, representation, patterns_per_sequence, 
                                             T_persistence=T_persistence, remove=remove)
        
        successes_collection[index, :] = aux['success']
        

    results_dic = {'manager': manager, 'sigma_vector': sigma_vector, 'success': successes_collection}

    return results_dic
```

#### Evaluate disambiguation success

```python
def evaluate_disambiguation_success(end, wrong_end, pattern_sequence):
    win = 0  # You already lost
    
    for x in pattern_sequence:
        if x == wrong_end:
            win = 0.0
            break
        if x == end:
            win = 1.0
            break
            
    return win
```

#### Calculate success vs sigma

```python
def calculate_success_vs_sigma(representation_overlap, sequence_overlap, sigma_number, trials, sigma_max,
                               T_persistence=0.050, remove=remove, cue_times=3):

    strict_maximum = True

    g_a = 1.0
    g_I = 2.0
    tau_a = 0.250
    G = 1.0
    sigma_out = 2.5
    tau_s = 0.010
    tau_z_pre = 0.025
    tau_z_post = 0.025

    hypercolumns = 10
    minicolumns = 20
    n_patterns = minicolumns
    patterns_per_sequence = int(n_patterns / 2)
    representation_overlap = representation_overlap
    sequence_overlap = sequence_overlap

    # Training protocol
    training_times_base = 0.100
    training_times = [training_times_base for i in range(n_patterns)]
    ipi_base = 0.0
    inter_pulse_intervals = [ipi_base for i in range(n_patterns)]
    inter_sequence_interval = 0.0
    resting_time = 1.0
    epochs = 1
    T_persistence = 0.100

    # Manager properties
    values_to_save = ['o']


    # Neural Network
    nn = Network(hypercolumns, minicolumns, G=G, tau_s=tau_s, tau_z_pre=tau_z_pre, tau_z_post=tau_z_post,
                     tau_a=tau_a, g_a=g_a, g_I=g_I, sigma_out=sigma_out, epsilon=epsilon, prng=np.random,
                     strict_maximum=strict_maximum, perfect=False, normalized_currents=True)


    # Build the manager
    manager = NetworkManager(nn=nn, dt=dt, values_to_save=values_to_save)
    # Build the representation
    activity_representation = create_overalaped_representation(manager, representation_overlap, sequence_overlap)

    representation = PatternsRepresentation(activity_representation, minicolumns=minicolumns)
    inter_pulse_intervals[9] = 1.0
    # Build the protocol
    protocol = Protocol()
    protocol.simple_protocol(representation, training_times=training_times, inter_pulse_intervals=inter_pulse_intervals,
                        inter_sequence_interval=inter_sequence_interval, epochs=epochs, resting_time=resting_time)

    # Run the protocol
    timed_input = manager.run_network_protocol_offline(protocol=protocol)
    
    results_dic = run_sigma_sweep(sigma_number, trials, sigma_max, manager, representation, patterns_per_sequence, 
                                  T_persistence=T_persistence, remove=remove, cue_times=cue_times)
    
    return results_dic
```

#### Plot success vs sigma

```python
def plot_success_vs_sigma(results_dic, index=0, label=' ', title=' ', ax=None):
    if ax is None:
        fig = plt.figure()
        ax = fig.add_subplot(111)
        
    y = np.mean(results_dic['success'], axis=1)
    x = results_dic['sigma_vector']

    trials = results_dic['success'].shape[1]
    xlabel = r'$\sigma$'
    n_samples = trials
    ci_low, ci_upp = sm.stats.proportion_confint(y * n_samples, n_samples, alpha=0.05, method='normal')

    ax.plot(x, y, 'o-', color=colors[index], lw=lw, ms=ms, label=label)

    # Plot the interval
    low = ci_low
    low[low < 0.0] = 0.0
    high = ci_upp
    ax.fill_between(x, low, high, color=colors[index], alpha=alpha_graph)

    ax.set_xlabel(xlabel)
    ax.set_ylabel('Success')
    ax.set_title(title)

    ax.set_ylim([-0.1, 1.1])

    ax.axhline(0, ls='--', color='gray')
    ax.axvline(0, ls='--', color='gray')
    
    ax.legend(frameon=False)
    
    return ax
```

### P50 functions

#### Number of iterations

```python
def number_of_iterations(start_size, desired_size=0.1):
    
    iterations = int(np.log(desired_size/ start_size) / np.log(0.5)) + 1

    return iterations
```

#### Find root

```python
def find_root(manager, representation, patterns_per_sequence, p_root, total_length, 
              n_samples, T_persistence, factor=1.0, verbose=False, remove=remove):
    
    final_size = 0.1
    iterations = number_of_iterations(start_size=total_length, desired_size=final_size)
    left = 0.0
    right = total_length
    tolerance = factor * np.sqrt(p_root * (1 - p_root) / n_samples)
    if verbose:
        print('number of iterations', iterations)
        print('left = ', left, 'right = ', right)
        print('************')
    for iteration in range(iterations):
        sigma_test = (left + right) * 0.5
        manager.nn.sigma_out = sigma_test
        manager.nn.sigma_in = sigma_test * np.sqrt(2 / manager.nn.tau_s)
        aux = calculate_total_recall_trials(n_samples, manager, representation, patterns_per_sequence, 
                                   T_persistence=T_persistence, remove=remove)
        result = np.mean(aux['success'])
        # First we check whether we are closer than the standar deviation for bernoulli estimates. 
        tol = np.abs(result - p_root) / tolerance
        
        if verbose:
            print('number of iteration : ', iteration)
            print('left = ', left, ' right = ', right, 'length = ', right - left)
            print('sigma test =', sigma_test, ' result =', result, ' p_root = ', p_root)
            print('tol = ', tol, ' factor  * std = ', tolerance)
            print('----')
            
        if tol < 1:
            break
        elif result > p_root:
            left = sigma_test
        elif result <= p_root:
            right = sigma_test
    
    length = right - left

    return result, sigma_test, length 
```

### Plot root diagram

#### An example

```python
def find_p(representation_overlap, sequence_overlap, p_roots, total_length, n_samples, 
           T_persistence, factor=1.0, verbose=False, remove=remove):
    
    strict_maximum = True

    g_a = 1.0
    g_I = 2.0
    tau_a = 0.250
    G = 1.0
    sigma_out = 2.5
    tau_s = 0.010
    tau_z_pre = 0.025
    tau_z_post = 0.025

    hypercolumns = 10
    minicolumns = 20
    n_patterns = minicolumns
    patterns_per_sequence = int(n_patterns / 2)
    representation_overlap = representation_overlap
    sequence_overlap = sequence_overlap

    # Training protocol
    training_times_base = 0.100
    training_times = [training_times_base for i in range(n_patterns)]
    ipi_base = 0.0
    inter_pulse_intervals = [ipi_base for i in range(n_patterns)]
    inter_sequence_interval = 0.0
    resting_time = 1.0
    epochs = 1
    T_persistence = 0.100

    # Manager properties
    values_to_save = ['o']


    # Neural Network
    nn = Network(hypercolumns, minicolumns, G=G, tau_s=tau_s, tau_z_pre=tau_z_pre, tau_z_post=tau_z_post,
                     tau_a=tau_a, g_a=g_a, g_I=g_I, sigma_out=sigma_out, epsilon=epsilon, prng=np.random,
                     strict_maximum=strict_maximum, perfect=False, normalized_currents=True)


    # Build the manager
    manager = NetworkManager(nn=nn, dt=dt, values_to_save=values_to_save)
    # Build the representation
    activity_representation = create_overalaped_representation(manager, representation_overlap, sequence_overlap)

    representation = PatternsRepresentation(activity_representation, minicolumns=minicolumns)
    inter_pulse_intervals[9] = 1.0
    # Build the protocol
    protocol = Protocol()
    protocol.simple_protocol(representation, training_times=training_times, inter_pulse_intervals=inter_pulse_intervals,
                        inter_sequence_interval=inter_sequence_interval, epochs=epochs, resting_time=resting_time)

    # Run the protocol
    timed_input = manager.run_network_protocol_offline(protocol=protocol)
    
    # Find the sigmas 
    results_dic = {}
    
    for p_root in p_roots: 
        results_dic[p_root] = find_root(manager, representation, patterns_per_sequence, p_root, 
                                        total_length, n_samples, T_persistence, factor=factor, 
                                        verbose=verbose, remove=remove)
    return results_dic
```

```python
n_samples = 10
p_roots = [0.5]
total_length = 15
representation_overlap = 1.0
sequence_overlap = 0.9
T_persistence = 0.100

aux = find_p(representation_overlap, sequence_overlap, p_roots, total_length, n_samples, 
             T_persistence, factor=1.0, verbose=True, remove=0.010)

for p_root in p_roots:
    p50, sigma50, length = aux[p_root]
    print('p50', p50, 'sigma50', sigma50, 'len', length)
```

```python
%%time
n_samples = 100
p_root = 0.5
total_length = 15
T_persistence = 0.100

representation_overlap_set = np.linspace(0.0, 1.0, num=11) 
sequence_overlap_set = np.linspace(0.0, 1.0, num=11) 

p50_matrix = np.zeros((11, 11))
sigma50_matrix = np.zeros((11, 11))
length_matrix = np.zeros((11, 11))

for index1, representation_overlap in enumerate(representation_overlap_set):
    print('-----------')
    print(representation_overlap)
    for index2, sequence_overlap in enumerate(sequence_overlap_set):
        aux = find_p(representation_overlap, sequence_overlap, p_roots, total_length, n_samples, 
                     T_persistence, factor=1.0, verbose=True, remove=0.010)

        p50, sigma50, length = aux[p_root]
        p50_matrix[index1, index2] = p50
        sigma50_matrix[index1, index2] = sigma50
        length_matrix[index1, index2] = length
```
