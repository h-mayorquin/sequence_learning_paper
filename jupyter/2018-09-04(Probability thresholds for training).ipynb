#### Sigma sweep

```python
def run_sigma_sweep(sigma_number, samples, sigma_max, manager, T_persistence=0.050):

    manager.set_persistent_time_with_adaptation_gain(T_persistence=T_persistence)
    tau_s = manager.nn.tau_s
    T_recall = T_persistence * minicolumns
    T_cue = 3 * tau_s
    nr = manager.canonical_network_representation

    sigma_vector = np.linspace(0, sigma_max, num=sigma_number)

    successes_collection = np.zeros((sigma_number, samples))
    persistent_times_collection = {index: [] for index in range(sigma_number)}
    sequence_collection = {index: [] for index in range(sigma_number)}

    for index, sigma in enumerate(sigma_vector):
        manager.nn.sigma_out = sigma
        manager.nn.sigma_in = sigma * np.sqrt(2 / tau_s)
        for sample in range(samples):
            aux = calculate_recall_quantities(manager, nr, T_recall, T_cue, remove=remove,
                                              reset=True, empty_history=True)
            success, pattern_sequence, persistent_times, timings = aux
            successes_collection[index, sample] = success
            persistent_times_collection[index].append(persistent_times[:n_patterns])
            sequence_collection[index].append(pattern_sequence)

    results_dic = {'manager': manager, 'sigma_vector': sigma_vector, 'success': successes_collection,
                   'persistent times': persistent_times_collection, 'sequences': sequence_collection,
                   'T_persistence': T_persistence}

    return results_dic
```

#### Plot mean success

```python
def plot_mean_success_vs_sigma(results_dictionary, n_samples, index=0, label=0, ax=None):

    if ax is None:
        fig = plt.figure()
        ax = fig.add_subplot(111)

    successes_collection = results_dictionary['success']
    sigma_vector = results_dictionary['sigma_vector']

    std = successes_collection.std(axis=1)
    mean_success = successes_collection.mean(axis=1)
    ci_low, ci_upp = sm.stats.proportion_confint(mean_success * n_samples, n_samples, alpha=0.05, method='normal')

    # Plot the mean
    ax.plot(sigma_vector, mean_success, 'o-', lw=lw, ms=ms, color=colors[index], label=label)
    
    if False:
        # Plot sigma05
        sigma09_arg = np.where(mean_success <= 0.9)[0]
        if sigma09_arg.size > 0:
            sigma09 = sigma_vector[sigma09_arg[0]]
            ax.plot(sigma09, 0.9, '*', ms=(ms + 20), color='black')

        sigma01_arg = np.where(mean_success <= 0.1)[0]
        if sigma01_arg.size > 0:
            sigma01 = sigma_vector[sigma01_arg[0]]
            ax.plot(sigma01, 0.1, '*', ms=(ms + 20), color='black')
    
    # Plot the interval
    low = ci_low
    low[low < 0.0] = 0.0
    high = ci_upp
    ax.fill_between(sigma_vector, low, high, color=colors[index], alpha=alpha_graph)

    ax.axhline(0, ls='--', color='gray')
    ax.axvline(0, ls='--', color='gray')
    ax.set_xlabel(r'$\sigma$')
    ax.set_ylabel('Success')
    ax.legend(frameon=False)

    return ax
```

#### Find success rate

```python
def find_success_rate(manager, n_samples, sigma_out, T_persistence):
    manager.nn.sigma_in = sigma_out * np.sqrt(2 / tau_s)
    manager.nn.sigma_out = sigma_out
    T_recall = T_persistence * minicolumns
    T_cue = manager.nn.tau_s
    I_cue = 0
    nr = manager.canonical_network_representation
    success_list = []
    for index in range(n_samples):
        aux = calculate_recall_quantities(manager, nr, T_recall, T_cue, remove=remove, reset=True, empty_history=True)
        success, pattern_sequence, persistent_times, timings = aux
        success_list.append(success)

    return np.mean(success_list)
```

#### Find root

```python
def find_root(manager, p_root, total_length, n_samples, T_persistence, factor=1.0, verbose=False):
    final_size = 0.1
    iterations = number_of_iterations(start_size=total_length, desired_size=final_size)
    left = 0.0
    right = total_length
    tolerance = factor * np.sqrt(p_root * (1 - p_root) / n_samples)
    if verbose:
        print('number of iterations', iterations)
        print('left = ', left, 'right = ', right)
        print('************')
    for iteration in range(iterations):
        sigma_test = (left + right) * 0.5
        result = find_success_rate(manager, n_samples, sigma_test, T_persistence)
        # First we check whether we are closer than the standar deviation for bernoulli estimates. 
        tol = np.abs(result - p_root) / tolerance
        
        if verbose:
            print('number of iteration : ', iteration)
            print('left = ', left, ' right = ', right, 'length = ', right - left)
            print('sigma test =', sigma_test, ' result =', result, ' p_root = ', p_root)
            print('tol = ', tol, ' factor  * std = ', tolerance)
            print('----')
            
        if tol < 1:
            break
        elif result > p_root:
            left = sigma_test
        elif result <= p_root:
            right = sigma_test
    
    length = right - left

    return result, sigma_test, length 
```

#### Find ps

```python
def find_p(training_time, ipi, tau_z_pre, minicolumns, hypercolumns, T_persistence, tau_z_post=0.005,
           n_samples=100, factor=1.5, total_length=50, p_roots=[0.5], verbose=False):
    
    strict_maximum = True

    g_a = 1.0
    g_I = 2.0
    tau_a = 0.250
    G = 1.0
    sigma_out = 0.0
    tau_s = 0.010
    tau_z_pre = tau_z_pre
    tau_z_post = tau_z_post

    hypercolumns = hypercolumns
    minicolumns = minicolumns
    n_patterns = minicolumns

    # Training protocol
    training_time_base = training_time
    training_times = [training_time_base for i in range(n_patterns)]
    inter_pulse_interval_base = ipi
    inter_pulse_intervals = [inter_pulse_interval_base for i in range(n_patterns)]
    inter_sequence_interval = 0.0
    resting_time = 2.0
    epochs = 1
    T_persistence = T_persistence

    # Manager properties
    dt = 0.001
    values_to_save = ['o']


    # Neural Network
    nn = Network(hypercolumns, minicolumns, G=G, tau_s=tau_s, tau_z_pre=tau_z_pre, tau_z_post=tau_z_post,
                 tau_a=tau_a, g_a=g_a, g_I=g_I, sigma_out=sigma_out, epsilon=epsilon, prng=np.random,
                 strict_maximum=strict_maximum, perfect=False, normalized_currents=True)


    # Build the manager
    manager = NetworkManager(nn=nn, dt=dt, values_to_save=values_to_save)
    # Build the representation
    representation = PatternsRepresentation(manager.canonical_activity_representation[:n_patterns],
                                            minicolumns=minicolumns)

    # Build the protocol
    protocol = Protocol()
    protocol.simple_protocol(representation, training_times=training_times, inter_pulse_intervals=inter_pulse_intervals, inter_sequence_interval=inter_sequence_interval, epochs=epochs, resting_time=resting_time)

    # Run the protocol
    timed_input = manager.run_network_protocol_offline(protocol=protocol)
    manager.set_persistent_time_with_adaptation_gain(T_persistence=T_persistence)


    # Find the sigmas 
    results_dic = {}
    
    for p in p_roots: 
        results_dic[p] = find_root(manager, p, total_length, n_samples, T_persistence, 
                                   factor=factor, verbose=verbose)

    return results_dic
```

#### Plot sigma estimates

```python
def plot_sigma_estimates(results_dictionary, title=' ', ax=None, legend=True, bars=True):

    if ax is None:
        fig = plt.figure()
        ax = fig.add_subplot(111)

    x_vector = results_dictionary['input']
    xlabel = results_dictionary['label']

    p_labels = results_dictionary['p_labels']
    p_roots = results_dictionary['p_roots']

    # Sigmas
    for index, (p_root, p_label) in enumerate(zip(p_roots, p_labels)):
        stored_dictionary = results_dictionary[p_root]
        sigma_vector = stored_dictionary['sigma']
        length_vector = stored_dictionary['length']
        label = r'$\sigma_{' + p_label + '}$'
        ax.plot(x_vector, sigma_vector, 'o-', lw=lw, ms=ms, color=colors[index], label=label)
        
        if bars:
            up = sigma_vector + length_vector * 0.5
            down = sigma_vector - length_vector * 0.5
            down[down < 0.0] = 0.0
            ax.fill_between(x_vector, up, down, color=colors[index], alpha=alpha_graph)

    ax.set_xlabel(xlabel)
    ax.set_ylabel(r'$\sigma$')

    ax.axhline(0, ls='--', color='gray')
    ax.axvline(0, ls='--', color='gray')

    if legend:
        ax.legend(frameon=False)
        ax.set_title(title)
    
    return ax
```

#### Plot p estimates

```python
def plot_p_estimates(results_dictionary, title=' ', ax=None, legend=True):
    
    if ax is None:
        fig = plt.figure()
        ax = fig.add_subplot(111)
    
    x_vector = results_dictionary['input']
    xlabel = results_dictionary['label']

    p_labels = results_dictionary['p_labels']
    p_roots = results_dictionary['p_roots']

    # Sigmas
    for index, (p_root, p_label) in enumerate(zip(p_roots, p_labels)):
        stored_dictionary = results_dictionary[p_root]
        p_vector = stored_dictionary['result']
        length_vector = stored_dictionary['length']
        label = r'$p_{' + p_label + '}$'
        ax.plot(x_vector, p_vector, 'o-', lw=lw, ms=ms, color=colors[index], label=label)
    

    ax.set_xlabel(xlabel)
    ax.set_ylabel(r'$p_{root}$')

    ax.set_ylim([-0.1, 1.0])
    ax.axhline(0, ls='--', color='gray')
    ax.axvline(0, ls='--', color='gray')
    
    for p_root in p_roots:
        ax.axhline(p_root, ls='--', color='gray')
    
    if legend:
        ax.legend(frameon=False)
        ax.set_title(title)
    
    return ax
```

```python
%%time

strict_maximum = True

g_a = 1.0
g_I = 2.0
tau_a = 0.250
G = 1.0
sigma_out = 0.0
tau_s = 0.010
tau_z_pre = 0.025
tau_z_post = 0.005

hypercolumns = 1
minicolumns = 5
n_patterns = minicolumns

# Training protocol
training_time_base = 0.100
training_times = [training_time_base for i in range(n_patterns)]
inter_pulse_interval_base = 0.0
inter_pulse_intervals = [inter_pulse_interval_base for i in range(n_patterns)]
inter_sequence_interval = 0.0
resting_time = 0.0
epochs = 1
T_persistence = 0.050

# Manager properties
dt = 0.001
values_to_save = ['o']

# Neural Network
nn = Network(hypercolumns, minicolumns, G=G, tau_s=tau_s, tau_z_pre=tau_z_pre, tau_z_post=tau_z_post,
             tau_a=tau_a, g_a=g_a, g_I=g_I, sigma_out=sigma_out, epsilon=epsilon, prng=np.random,
             strict_maximum=strict_maximum, perfect=False, normalized_currents=True)


# Build the manager
manager = NetworkManager(nn=nn, dt=dt, values_to_save=values_to_save)
# Build the representation
representation = PatternsRepresentation(manager.canonical_activity_representation[:n_patterns],
                                        minicolumns=minicolumns)

training_times_set = [0.050, 0.200]
sigma_number = 40
n_samples_here = n_samples
sigma_max = total_length
results_collection = {}


for training_times in training_times_set:
    print(training_times)
    # Build the protocol
    protocol = Protocol()
    protocol.simple_protocol(representation, training_times=training_times, 
                             inter_pulse_intervals=inter_pulse_intervals, inter_sequence_interval=inter_sequence_interval, epochs=epochs, resting_time=resting_time)

    # Run the protocol
    timed_input = manager.run_network_protocol_offline(protocol=protocol)

    results_dictionary = run_sigma_sweep(sigma_number, n_samples_here, sigma_max, manager, 
                                                T_persistence=T_persistence)
    for p_root in p_roots:
        aux = find_root(manager, p_root, total_length, n_samples_here, 
                        T_persistence=T_persistence, factor=factor, verbose=False)
        results_dictionary[p_root] = aux
        
        
    results_collection[training_times] = results_dictionary
```
