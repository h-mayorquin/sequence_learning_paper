## Evolution of weights
Here we compared how the on-line rule compared with the offline rule. It seems to be very similar. One important thing to say is that if you are learning all the time (even in the inter-sequence-interval) then you may get the statistics wrong because you may get some initial disturbance in the statistics during the first group of inter-sequence-intervals that will get replicated again in the system (because the system replicates the statistics). This was done with inter-sequence-interval = sequence_interval maybe it works just fine with smaller inter-sequence-intervals.

We will run first an example with the off line learning
